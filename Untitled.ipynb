```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
```

```python

# Load the VGG16 model pre-trained on ImageNet
model = VGG16(weights='imagenet', include_top=False)
```

```python
# Display the model summary
model.summary()
```

```python
# Print the names of all layers in the model
for layer in model.layers:
    print(layer.name)
```

```python



# Specify the layer to visualize
layer_name = 'block1_conv1'  # Change this to the layer you want to visualize

# Create a new model that outputs the feature maps of the specified layer
layer_output = model.get_layer(layer_name).output
model1 = Model(inputs=model.input, outputs=layer_output)


```

```python
# Load and preprocess an image
img_path = 'C:/Users/praba/zennit-crp/tutorials/images/dog1.jpg'  
img = tf.keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
img = tf.keras.preprocessing.image.img_to_array(img)
img = np.expand_dims(img, axis=0)
img = tf.keras.applications.vgg16.preprocess_input(img)

# Get the feature maps
feature_maps = model1.predict(img)

# Plot the feature maps
num_filters = feature_maps.shape[-1]
size = feature_maps.shape[1]


# Create a custom colormap that transitions from white to red
colors = [(1, 1, 1), (1, 0, 0)]  # white to red
n_bins = 100  # Number of bins for the colormap
custom_cmap = LinearSegmentedColormap.from_list('white_to_red', colors, N=n_bins)


fig, ax = plt.subplots(nrows=num_filters // 8, ncols=8, figsize=(15, 15))
for i in range(num_filters):
    row = i // 8
    col = i % 8
    ax[row, col].imshow(feature_maps[0, :, :, i], cmap=custom_cmap)
    ax[row, col].axis('off')
plt.savefig('C:/Users/praba/Desktop/z_img/feature_maps_1.png', bbox_inches='tight')

plt.show()
```
